## Part 1
- Ship starts facing east
- What is sum of EW/NS abs distance at end

```python
dir_dict = {'N':0, 'E':1, 'S':2, 'W':3}
amt_dict = {90:1, 180:2, 270:3, 360:0}
```

```python
def adj_dir(command, amount, current_dir):
    if command == 'R':
        direction = 1
    else:
        direction = -1
    current_dir = (current_dir + direction*amt_dict.get(amount))%4
    return current_dir
```

```python
n,s,e,w = 0,0,0,0
current_dir = 1

for line in lines:
    command = line[0]
    amount = int(line[1:])
    if command == 'F':
        if current_dir == 0:
            n += amount
        if current_dir == 1:
            e += amount
        if current_dir == 2:
            s += amount
        if current_dir == 3:
            w += amount

    if command in ['L','R']:
        current_dir = adj_dir(command, amount, current_dir)
    else:
        if command == 'N':
            n += amount
        if command == 'S':
            s += amount
        if command == 'E':
            e += amount
        if command == 'W':
            w += amount
    
print (abs(n-s) + abs(e-w))
```

## Part 2


```python
def adj_dir_simp(command, amount, x,y):    
    if command == 'R':
        amount = 360 - amount
    
    if amount == 180:
        tempx = -x
        tempy = -y
    
    if amount == 90:
        tempy = x
        tempx = -y
    
    if amount == 270:
        tempy = -x
        tempx = y
       
    return tempx,tempy
```

```python
total_x, total_y = 0,0
x = 10
y = 1
x_total = 0
y_total = 0


for line in lines:
    command = line[0]
    amount = int(line[1:])
    if command == 'F':
        x_total += amount * x
        y_total += amount * y 
        
    if command in ['L','R']:
        x,y = adj_dir_simp(command, amount, x,y)
        
        
    else:
        if command == 'N':
            y += amount
        elif command == 'S':
            y -= amount
        elif command == 'E':
            x += amount
        elif command == 'W':
            x -= amount

print (abs(x_total) + abs(y_total))
```
